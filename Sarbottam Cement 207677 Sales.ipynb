Adding Titile and labels

```python
fig, axes = plt.subplots(figsize=(14,8))
quantity_plot = sales_quantity.plot(kind='bar' , title="Product Type vs Sale Quantity")
quantity_plot.set_xlabel('Product Code')
quantity_plot.set_ylabel('Quantity')
```
